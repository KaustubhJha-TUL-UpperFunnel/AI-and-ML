```python
import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data_new.csv")
```

```python
df.info()
```

```python
#inputs and outputs of perceptrons should be in matrix
X=df[['x','y']].as_matrix()
Y=df[['label']].as_matrix()
```

```python
p = np.random.permutation(X.shape[0])
#the values of x and y shuffle after this ,according to index value of p
X=X[p]
Y=Y[p]
```

```python





#define neurons/perceptrons
num_units = 1
num_inputs = 2
num_outputs = 1
#X[Value = None, input = 2]
x = tf.placeholder(tf.float32,[None,num_inputs])
#Y[Value = None, input = 1]
y = tf.placeholder(tf.float32,[None,num_outputs])
```

```python
#weights and bias
#tf.random_normal - Tensor type random values of weight
#weight -[w1,w2]
weight = tf.Variable(tf.random_normal([num_inputs,num_units]))
bias=tf.Variable(tf.random_normal([num_units]))
#activation_fns(XW+b)
#z=X.W + bias
#matmul-Matrix multiplication
output = tf.add(tf.matmul(x,weight),bias)
output = tf.sigmoid(output)
```

```python
sess = tf.Session()
```

```python
#define the cost function fns and the optimizer
#avg -cross entropy:-log(h(x)-log(1-h(x))
#cost function of neuron
loss =tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(labels=y,logits=output))
#model optimization with  alpha =0.1
gdo = tf.train.GradientDescentOptimizer(learning_rate=0.4).minimize(loss)
init = tf.global_variables_initializer()
```

```python
def get_batch(x,y,batch_size,idx):
    return x[batch_size*idx:batch_size*(idx +1)],y[batch_size*idx:batch_size*(idx+1)]
```

```python
#epoch is iterations for each batch size is sample from data 
epochs = 1000
batch_size = 10
total_batch_count = int(X.shape[0]/batch_size)
with tf.Session() as sess:
    sess.run(init)
    for e in range(epochs):
        avg_loss =0 
        for i in range(total_batch_count):
            #function will return X[0:20],Y[0:20]
            batch_x,batch_y = get_batch(X,Y,batch_size,i)
            #c will store the total loss values
            nonsenseuseless,c = sess.run([gdo,loss],feed_dict={x:batch_x,y:batch_y})
            avg_loss+=c/total_batch_count
        print("Epoch",e+1,'loss',avg_loss)
    #output from tf.output
    preds=output.eval({x:X[:4]})
    print(np.round(preds))
    sess.close()
```
